# Usando Seaborn

```python
np.random.seed(42)
n = 1000
pet_smokers = 0.2

flag_fumante = np.random.rand(n) < pet_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante})

dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'Não Fumante'})
```

```python
dados.shape
```

```python
dados.head()
```

```python
# Set the Seaborn style to 'ticks'
sns.set(style='ticks')

# Assuming you have a DataFrame named 'dados' with columns 'altura', 'peso', and 'flag_fumante'

# Create a scatterplot with a linear regression line
sns.lmplot(x='altura',
           y='peso',
           data=dados,
           hue='flag_fumante',
           palette=['tab:blue', 'tab:orange'],
           height=7)
           
# Set labels and title
plt.xlabel('altura (cm)')
plt.ylabel('peso (kg)')
plt.title("Relação entre altura e peso")

# Remove the top and right spines
sns.despine()

# Show the plot
plt.show()
```

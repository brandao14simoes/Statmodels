O coeficiente de correlao  uma medida estatstica que indica a fora e a direo da relao linear entre duas variveis numricas. Ele varia entre -1 e 1, onde:

Um coeficiente de correlao igual a 1 indica uma correlao linear perfeita positiva, ou seja, quando uma varivel aumenta, a outra varivel tambm aumenta na mesma proporo.

Um coeficiente de correlao igual a -1 indica uma correlao linear perfeita negativa, ou seja, quando uma varivel aumenta, a outra varivel diminui na mesma proporo.

Um coeficiente de correlao igual a 0 sugere que no h correlao linear entre as duas variveis.

Regresso Linear Simples

A Regresso Linear  uma tcnica estatstica utilizada para modelar a relao entre uma varivel dependente (tambm chamada de varivel resposta ou varivel alvo) e uma ou mais variveis independentes (tambm chamadas de variveis explicativas ou preditoras). 

```python
df_dsa.head()
```

```python
y = df_dsa["valor_aluguel"]
```

```python
x = df_dsa["area_m2"]
```

```python
x = sm.add_constant(x)
```

```python
modelo = sm.OLS(y, x)
```

```python
resultado = modelo.fit()
```

```python
print(resultado.summary())
```

```python
plt.figure(figsize = (12, 8))
plt.xlabel("area_m2", size = 16)
plt.ylabel("valor_aluguel", size = 16)
plt.plot(x["area_m2"], y, "o", label = "Dados Reais")
plt.plot(x["area_m2"], resultado.fittedvalues, "r-", label = "linha de Regrasso (Previses do Modelo)")
plt.legend(loc = 'best')
plt.show()
```
